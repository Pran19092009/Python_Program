```python
import numpy as np
```

```python
marks=np.random.randint(0,99,(50))
print("student marks",marks)
total=marks.sum()
average_marks=total/50

print("\nTotal of the student marks:",total)
print("\nAverage of the student marks:",average_marks)
print("\nThe highest marks:",np.max(marks))
print("\nThe lowes marks:",np.min(marks))

above_75 = np.sum(marks > 75)
below_35 = np.sum(marks < 35)
between_35_75 = np.sum((marks >= 35) & (marks <= 75))

print("\n70 above student makrs total =",above_75)
print("\n30 below student makrs total =",below_35)
print("\nbetween the 35-75 total =",between_35_75)

#5 Top and 5 Bottom student marks
top=np.sort(marks)[-5:]
bottom=np.sort(marks)[0:6]
print("\nTop 5 student mark:",top)
print("\nBottom 5 student mark:",bottom)

#median and standard deviation
print("\nMedian of the marks:",np.median(marks))
print("\nStandard Deviation of the marks",np.std(marks))

####################################
# 5 subject marks
sub_marks=np.random.randint(35,101,size=(50,5))
print("\nEach Student 5 Subject Marks:\n",sub_marks)
                    
Total=np.sum(sub_marks,axis=1)
Average=np.mean(sub_marks, axis=1)
Percentage= (Total / 500) * 100

print("5 Subject Total Of Each Student:")
for i in range(50):
    print(f"{i+1} = \t{Total[i]}")
    

topper=np.argmax(sub_marks,0)
subjects=["Marathi","HIndhi","English","Maths","Social Science"]
for i in range(5):
    print(f"{subjects[i]} : Student {topper[i]+1} Scored {sub_marks[topper[i],i]} marks")
print("=====================REPORT CARD==========================")
print("Student\tTotal\tAverage\t\tPercentage\tGrade")

grade = []

for i in range(50):

    if Percentage[i] >= 90:
        grade.append("A+")
    elif Percentage[i] >= 80:
        grade.append("A")
    elif Percentage[i] >= 70:
        grade.append("B")
    elif Percentage[i] >= 60:
        grade.append("C")
    elif Percentage[i] >= 40:
        grade.append("D")
    else:
        grade.append("F")

    print(f"{i+1}\t{Total[i]}\t{Average[i]:2f}\t{Percentage[i]:.2f}%\t\t{grade[i]}")
```
