```python
above_90 = 0
below_35 = 0

f = open("students.txt", "r")
for line in f:
    name, roll, marks = line.strip().split(",")
    if int(marks) > 90:
        above_90 = above_90 + 1
        
    if int(marks) < 35:
        below_35 = below_35 + 1
f.close()

print("Students Scoring More Than 90 :", above_90)
print("Students Scoring Less Than 35 :", below_35)
```

```python
while True:

    print("\n===== STUDENT RECORD SYSTEM =====")
    print("1. Add New Student Record")
    print("2. View Student record ")
    print("3. Search Student")
    print("4. Update Marks")
    print("5. Delete Student")
    print("6. Exit")

    choice = input("Enter Choice: ")

    # ============== ADD ==============
    if choice == "1":

        name = input("Enter Student Name: ")
        roll = input("Enter Student Roll No: ")
        marks = input("Enter Student Marks: ")

        with open("data.txt", "a") as file:
            file.write(name + "," + roll + "," + marks +"\n")

        print("Student Added Successfully")

    # ============== VIEW ==============
    elif choice == "2":

        with open("data.txt", "r") as file:
            data = file.readlines()
            if len(data) == 0:
                print("No Records Found")
            else:
                print("\nStudent Records")
                for line in data:
                    print(line.strip())

    # ============== SEARCH ==============
    elif choice == "3":

        roll = input("Enter Roll Number: ")
        found = False

        with open("data.txt", "r") as file:
            for line in file:
                name, r, marks= line.strip().split(",")
                if roll == r:
                    print("\nRecord Found")
                    print("Name :", name)
                    print("Roll :", r)
                    print("Marks:", marks)
                    found = True
                    break
        if found == False:
            print("Student Not Found")

    # ============== UPDATE ==============
    elif choice == "4":

        roll = input("Enter Roll Number: ")
        new_marks = input("Enter New Marks: ")
        records = []
        with open("data.txt", "r") as file:
            for line in file:
                name, r, marks= line.strip().split(",")
                if roll == r:
                    records.append(name + "," + r + "," + new_marks +"\n")
                else:
                    records.append(line)
        with open("data", "w") as file:
            file.writelines(records)
        print("Marks Updated Successfully")

    # ============== DELETE ===================
    elif choice == "5":

        roll = input("Enter Roll Number: ")
        records = []
        with open("data.txt", "r") as file:
            for line in file:
                name, r, marks = line.strip().split(",")

                if roll != r:
                    records.append(line)
        with open("data.txt", "w") as file:
            file.writelines(records)
        print("Record Deleted Successfully")

    # ============== EXIT ==============
    elif choice == "6":
        print("Thank You")
        break

    else:
        print("Invalid Choice")
```
